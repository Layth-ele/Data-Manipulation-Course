```python
# Importing libraries

import numpy as np
import pandas as pd

import os

# os.getcwd()
```

```python
# Loading dataset

fileName = 'survey_results_public.csv'
schemaName = 'survey_results_schema.csv'


df = pd.read_csv(fileName, index_col='ResponseId') 
schema_df = pd.read_csv(schemaName,index_col='qname')
```

```python
# Filter a counties with high salarirs

high_salary = (df['ConvertedCompYearly'] > 70000)
```

```python
df.loc[high_salary,['Country','LanguageHaveWorkedWith','ConvertedCompYearly']]
```

```python
# Filtering by specific country

countries = ['United States of America','Sweden','Russian Federation']

filt1 = df['Country'].isin(countries)
df.loc[filt1]
```
