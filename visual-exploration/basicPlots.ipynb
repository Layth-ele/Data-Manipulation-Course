# Bar Plots
<h3>Comparing discrete data. Available through either pandas or matplotlib.</h3>

```python
chest_pain = df.groupby('cp').median().reset_index()
chest_pain
```

```python
#Using pandas library to show the bar

chest_pain.plot.bar(x ='cp',y='age',figsize = (12,6));
```

```python
chest_pain.plot.bar(x ='cp',figsize = (12,6));
```

```python
# Using matplotlib library to show the bar

fig, ax = plt. subplots()
ax.bar (chest_pain["cp"], chest_pain["age"], label="age")
ax.set_xlabel ("cp")
ax.legend();
```

```python
# Styling

    fig, ax = plt. subplots()
    ax.bar (chest_pain["cp"], chest_pain["age"], label="age",color =['r','g','b','k'])
    ax.set_xlabel ("cp")
    ax.legend();
```

```python
# Using bar in seaborn

ax = sns.barplot(data= df, x= 'cp', y='age',errcolor='k')
```

# Scatter Plot

```python
# Using pandas scatter plot

df.plot.scatter(x= 'age',y= 'trestbps',label="age")
```

```python
# Using matplotlab scatter

fig, ax = plt. subplots()
ax.scatter(df["age"], df["trestbps"], label="age" ,marker='^',\
           edgecolors= "k" ,s=60 ,c=df["age"], alpha=.6)
ax.set_xlabel("age")
ax.set_ylabel ("trestbps")
ax.legend();
```

```python
# Using seaborn scatter

sns.scatterplot(data = df, x= 'age', y=  'trestbps',hue= 'age', s=60)
```

# Line Plot

```python
ages = df.groupby('age').median().reset_index()

ages.head()
```

```python
# Using pandas line plot

ages.plot.line('age','cp');
ages.plot.line('age',['cp','chol','trestbps','thalach']);
```

```python
# Using line plot with matplotlib

fig, ax = plt. subplots()
ax.plot(ages["age"], ages["trestbps"], label="age" ,ls= ':',lw= 1.8 )
ax.set_xlabel("age")
ax.set_ylabel ("trestbps")
ax.legend();
```

```python
# Using line plot with seaborn

sns.lineplot(data = ages , x= 'age' ,y= 'trestbps');
```
