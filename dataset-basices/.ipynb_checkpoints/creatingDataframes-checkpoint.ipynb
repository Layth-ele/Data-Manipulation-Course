# Saving and Serialising the DataFrames


```python
# Imorting libraries

import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1- Array

data = np.random.random(size= (5,3))

df = pd.DataFrame(data,columns = ['A','B','C'])
df
```

```python
#3- Structured data
dtype = [("A", np.int), ("B", (np.str,20)),('C', np.bool)]

data = np.array([(1,'Sam',1),(2,'Layth',0),(3,'Kori',1),(4,'Lina',1)],dtype=dtype)
df = pd.DataFrame(data)
df

```
